```python
def PreGenerateRequest1(scrap_res,start_date,end_date,EntitiesFromKG=True):
    
    if EntitiesFromKG==True:
        # Keep most Similar description between the kowledge graph descriptions and the companies given descriptions
        result=Filtration(scrap_res,KeepMostSimilarDescription=True)
        #list of entities_ids (that will be used in the request later)
        listentities_ids=list(result['item_id'])
        listentities_ids=list(filter(("").__ne__, listentities_ids))
        #Create mrequest with all item_ids found in kg from the companies of the csv file
        mrequest=availableEntitiesProperties(listentities_ids)
    elif EntitiesFromKG==True:
        entities_df=CreateSearchedCompanies(scrap_res)
        #Instantiate the "entities" part of the request
        mrequest = {}
        entities_df = entities_df.to_json(orient='records')
        mrequest['entities'] = json.loads(entities_df)
    concept_risk_file = "data/concepts_risks.json"
    concepts_risks = load_request_from_file(concept_risk_file)
    quality_score = 50
    min_match = 1
    min_repeat = 1
    similarity_threshold = 0.7
    precompute = False
    workers = ["quality-score", "concept","ner-linking", "entity-similarity","embedder-indicators"]
    sites_exclude = ["pastebin.com"]

    languages = ["english"]
    mrequest['workers'] = workers
    mrequest['start_date'] = start_date
    mrequest['end_date'] = end_date
    mrequest['qscore'] = quality_score
    mrequest['sites_exclude'] = sites_exclude
    mrequest['concepts'] = concepts_risks
    mrequest['languages'] = languages
    mrequest['similarity_threshold'] = similarity_threshold
    mrequest['languages'] = languages
    mrequest['min_match'] = min_repeat
    mrequest['min_repeat'] = min_match
    mrequest['precompute'] = precompute
    try:
        del mrequest['not_found']
    except:
        pass
    return mrequest
        
```
